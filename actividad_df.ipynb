```python
import pandas as pd
import sqlite3

#Aqui agrega la conexión a la base de datos
conn = sqlite3.connect("Salaries.sqlite")

#Crea el dataframe con los datos del datasets
df = pd.read_sql_query("SELECT * FROM Salaries", conn)

conn.close()
df.head(5)
```

1. **Top 10 empleados con mayor salario total (incluyendo beneficios)**
Extrae e imprime un top 10 de expleados con mayor salario incluyendo los beneficios 


```python
#Se ordenan los 10 valores más altos de "TotalPayBenefits" de forma descendente
mayor_salario = df.sort_values(by="TotalPayBenefits", ascending=False).head(10)
mayor_salario
```

2. **Filtrar empleados con más de 50,000 en horas extra, la columna a tratar es: (OvertimePay)**

```python
#En la columna "OvertimePay" hay ciertos valores que dicen "Not Provided", si no se convierte esto a NaN no se va a poder trabajar con datos númericos.
#Lo demas se convierte a int
df["OvertimePay"] = pd.to_numeric(df["OvertimePay"], errors="coerce").fillna(0).astype(int)

df[df["OvertimePay"] > 50000]
```

3. **Contar cuántos empleados únicos hay por año**

```python
#Se agrupa el nombre de los empleados por año y se cuentan sin repetición
empleados_por_año = df.groupby("Year")["EmployeeName"].nunique()
empleados_por_año
```

4. **Ver cuántos cargos únicos (JobTitle) existen y los 5 más comunes**

```python
#Conteo de cargos únicos evitando que se repitan
cargos_unicos = df["JobTitle"].nunique()
print("Cargos únicos: ", cargos_unicos)

#Se realiza el conteo de los 5 más comunes
cargos_comunes = df["JobTitle"].value_counts().head(5)
print("\nLos 5 cargos más comunes: ")
print(cargos_comunes)
```

5. **Mostrar el salario total (TotalPay) promedio por año**

```python
#Se utiliza el método mean para calcular el promedio de la paga total por año
salario_promedio = df.groupby("Year")["TotalPay"].mean()
salario_promedio
```

6. **Exporta todos los datos a un CSV y muestra los ultimos 10 valores**

```python
#Se realiza la conversión a CSV, sin el indice
df.to_csv("Salaries.csv", index=False)

ultimos_valores = df.tail(10)
ultimos_valores
```
